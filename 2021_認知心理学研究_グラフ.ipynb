```python
##参考にしたWebサイト##
##軸周り　https://www.yutaka-note.com/entry/matplotlib_axis
##枠線消し　https://qiita.com/irs/items/fe909442be057f0efb48
##折れ線　https://3pysci.com/python-matplotlib-3/
##エラーバー　https://3pysci.com/python-matplotlib-6/#2
##目盛りフォントサイズ　https://www.delftstack.com/ja/howto/matplotlib/how-to-set-tick-labels-font-size-in-matplotlib
##凡例　https://self-development.info/matplotlib%E3%81%A7%E6%A3%92%E3%82%B0%E3%83%A9%E3%83%95%E3%81%AE%E5%87%A1%E4%BE%8B%E3%82%92%E3%82%AB%E3%82%B9%E3%82%BF%E3%83%9E%E3%82%A4%E3%82%BA%E3%81%99%E3%82%8B%E6%96%B9%E6%B3%95%E3%80%90python/



import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.font_manager

##variables##
fontsize = 24
linewidth = 3
capsize = 4
legend_size = 24
marker = "o"
markersize = 8

#plt.rcParams["font.serif"] = "times"
#plt.rcParams['font.family'] = "TImes New Roman"
#plt.rcParams['font.family'] = 'serif'
#plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']

fig = plt.figure(figsize = (12, 6))
plt.subplots_adjust(wspace=0.25)

x = [0, 1, 2, 3, 4]

##data##################################
ACC_smile_list = [0.88, 0.61, 0.62, 0.53, 0.64]
ACC_sad_list = [0.82, 0.57, 0.47, 0.38, 0.35]

ACC_smile_sem_list = [0.05, 0.08, 0.04, 0.05, 0.09]
ACC_sad_sem_list = [0.03, 0.05, 0.05, 0.05, 0.07]

Res_smile_list = [0.17, 1.12, 1.99, 2.74, 3.52]
Res_sad_list = [0.26, 1.04, 1.69, 2.39, 3.00]

Res_smile_sem_list = [0.07, 0.06, 0.06, 0.10, 0.15]
Res_sad_sem_list = [0.07, 0.08, 0.10, 0.13, 0.19]
##dataここまで##################################

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(x, ACC_smile_list, linewidth = linewidth, marker = marker, markersize = markersize, color = "black", label = "Smile", zorder =10)
ax1.plot(x, ACC_sad_list, linewidth = linewidth, marker = marker, markersize = markersize, color = "darkgray", label ="Sad", zorder =10)
ax1.errorbar(x, ACC_smile_list, yerr= ACC_smile_sem_list, linestyle = "None", capsize = capsize, ecolor="black")
ax1.errorbar(x, ACC_sad_list, yerr= ACC_sad_sem_list, linestyle = "None", capsize = capsize, ecolor="black")


ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(x, Res_smile_list, linewidth = linewidth, marker = marker, markersize = markersize, color = "black", label ="Smile", zorder =10)
ax2.plot(x, Res_sad_list, linewidth = linewidth, marker = marker, markersize = markersize, color = "darkgray", label ="Sad", zorder =10)
ax2.errorbar(x, Res_smile_list, yerr= Res_smile_sem_list, linestyle = "None", capsize = capsize, ecolor="black")
ax2.errorbar(x, Res_sad_list, yerr= Res_sad_sem_list, linestyle = "None", capsize = capsize, ecolor="black")

ax1.set_xlabel("Number of targets", size = fontsize, labelpad = 6)
ax1.set_ylabel("Accuracy", size = fontsize)
ax2.set_xlabel("Number of targets", size = fontsize, labelpad = 6)
ax2.set_ylabel("Mean enumerated number", size = fontsize)

##軸の最小値と最大値
ax1.set_ylim(0, 1)
ax2.set_ylim(0, 4)

##目盛り
plt.rcParams["xtick.direction"] = "in"
plt.rcParams["ytick.direction"] = "in"
ax1.set_xticks([0, 1, 2, 3, 4])
ax1.set_xticklabels([0, 1, 2, 3, 4], fontsize = fontsize)
ax1.set_yticks([0, 0.25, 0.5, 0.75, 1])
ax1.set_yticklabels([0, 0.25, 0.5, 0.75, 1], fontsize = fontsize)
ax2.set_xticks([0, 1, 2, 3, 4])
ax2.set_xticklabels([0, 1, 2, 3, 4], fontsize = fontsize)
ax2.set_yticks([0, 1, 2, 3, 4])
ax2.set_yticklabels([0, 1, 2, 3, 4], fontsize = fontsize)

##枠線消し
ax1.spines["right"].set_visible(False)
ax1.spines["top"].set_visible(False)
ax2.spines["right"].set_visible(False)
ax2.spines["top"].set_visible(False)

##グラフ（ax）タイトル
ax1.set_title("a", x= 0.03, y=1.03, fontsize = fontsize+3, fontweight = "bold")
ax2.set_title("b", x= 0.03, y=1.03, fontsize = fontsize+3, fontweight = "bold")

#ax1.set_title("a", x= -0.235, y=1.03, fontsize = fontsize+3, fontweight = "bold")
#ax2.set_title("b", x= -0.12, y=1.03, fontsize = fontsize+3, fontweight = "bold")

plt.legend(prop={"size": legend_size}, loc ="best", framealpha =0.0)
#plt.show()
plt.savefig("test_new.png", format ="png", dpi = 600, bbox_inches = "tight")
```
